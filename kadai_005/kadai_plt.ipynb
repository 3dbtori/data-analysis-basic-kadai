```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nrows = 2
ncols = 3
fig, axes = plt.subplots(nrows, ncols, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1)

x2 = np.random.normal(15, 5, 2000)
axes[0, 2].hist(x2)

x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 0].scatter(x3, y3)

plt.show()
```
